```python
def factors(n):
    for val in range(1, n+1):
        if n % val == 0:
            yield val

factors_of_20 = factors(20)

print(next(factors_of_20))
print(next(factors_of_20))
print(next(factors_of_20))
```

```python
n = 20
gen_exp = (val for val in range(1, n+1) if n % val == 0)

print(next(gen_exp))
print(next(gen_exp))
print(next(gen_exp))
```

```python
def yield_multiple_statements():
    yield "This is the first statement"
    yield "This is the second statement"
    yield "This is the third statement"
    yield "This is the last statement."

example = yield_multiple_statements()

print(next(example))
print(next(example))
print(next(example))
print(next(example))
```
