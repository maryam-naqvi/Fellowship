```python
# Week 1 Mini-Project: Python Basics

# Question 1: User Data Collector
def collect_user_data():
    user_data = {}
    user_data["name"] = input("Enter your name: ")
    user_data["age"] = input("Enter your age: ")
    user_data["email"] = input("Enter your email: ")
    user_data["favorite_number"] = input("Enter your favorite number: ")

    # Validate email (must contain '@' and '.')
    if "@" in user_data["email"] and "." in user_data["email"]:
        print("\nUser Data Collected Successfully:")
        print(user_data)
    else:
        print("Invalid email format. Please enter a valid email.")

collect_user_data()

# Question 2: Even or Odd?
def is_even(number):
    if number % 2 == 0:
        print(f"{number} is even.")
    else:
        print(f"{number} is odd.")

num = int(input("\nEnter a number to check even/odd: "))
is_even(num)

# Question 3: Temperature Converter 
def convert_temperature(temp, scale):
    if scale.upper() == "C":
        print(f"{temp}°C is equal to {(temp * 9/5) + 32}°F")  # Convert to Fahrenheit
    elif scale.upper() == "F":
        print(f"{temp}°F is equal to {(temp - 32) * 5/9}°C")  # Convert to Celsius
    else:
        print("Invalid scale! Use 'C' for Celsius or 'F' for Fahrenheit.")

temp = float(input("\nEnter temperature: "))
scale = input("Enter scale (C for Celsius, F for Fahrenheit): ")
convert_temperature(temp, scale)

# Question 4: Finding Min & Max
def find_max_min(numbers_list):
    return max(numbers_list), min(numbers_list)

numbers = []
print("\nEnter 5 numbers:")
for i in range(5):
    num = int(input(f"Number {i+1}: "))
    numbers.append(num)

max_value, min_value = find_max_min(numbers)
print(f"Maximum: {max_value}, Minimum: {min_value}")

# Question 5: Student Data Manager 
students = []
for i in range(3):
    name = input(f"\nEnter name of student {i+1}: ")
    age = input(f"Enter age of {name}: ")
    grade = input(f"Enter grade of {name}: ")
    students.append((name, age, grade))  # Store as tuple

# Convert list to dictionary using a basic loop 
students_dict = {}
for student in students:
    students_dict[student[0]] = (student[1], student[2])

print("\nStudent Data Dictionary:", students_dict)

# Question 6: Inventory Management System
def update_inventory(inventory_dict, item, quantity):
    if item in inventory_dict:
        inventory_dict[item] = max(0, inventory_dict[item] + quantity)  # Ensure no negative values
    else:
        inventory_dict[item] = max(0, quantity)
    return inventory_dict

# Initial inventory
inventory = {
    "Apples": 10,
    "Bananas": 5,
    "Oranges": 8,
    "Milk": 2,
    "Bread": 3
}

print("\nInitial Inventory:", inventory)

# Updating inventory for 3 items
for i in range(3):
    item = input("\nEnter item name to update: ")
    quantity = int(input(f"Enter quantity to add/remove (negative to remove) for {item}: "))
    inventory = update_inventory(inventory, item, quantity)

print("\nUpdated Inventory:", inventory)
```
